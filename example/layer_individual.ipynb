```python
from pyGCodeDecode import gcode_interpreter
import numpy as np
anisoprint_A4 = {
    # general properties
    "nozzle_diam": 0.4,
    "filament_diam": 1.75,
    # default settings
    "velocity": 35,
    "acceleration": 1000,
    "jerk": 10,
    # axis max speeds
    "vX": 180,
    "vY": 180,
    "vZ": 30,
    "vE": 33,
}
```

```python
simulation = gcode_interpreter.simulate(filename=r"stern.gcode",printer=anisoprint_A4)
```

```python
layer_list = list()
new_list = list()
for state in simulation.states:
    new_list.append(state)
    if state.comment is not None and "LAYER" in state.comment:
        layer_list.append(new_list)
        new_list = list()
```

```python
layertime = list()
for layer in layer_list:
    simulation.refresh(new_state_list=layer)
    layertime.append(simulation.blocklist[-1].get_segments()[-1].t_end)
```

```python
import matplotlib.pyplot as plt

```

```python
layertime
```

```python
plt.plot(layertime)
```
