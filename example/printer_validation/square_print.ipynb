```python
%matplotlib widget
from pyGCodeDecode import gcode_interpreter
import matplotlib.pyplot as plt
import numpy as np
# ipympl can be install via pip: "pip install ipympl"
#PRINTER DEFINITION
anisoprint_A4 = {
    #general properties
    "nozzle_diam"   :   0.4,
    "filament_diam" :   1.75,
    
    #default settings
    "velocity"      :   35,
    "acceleration"  :   1000,
    "jerk"          :   10,
    
    #axis max speeds
    "Vx"            :   180,
    "Vy"            :   180,
    "Vz"            :   30,
    "Ve"            :   33
    }
```

```python
initial_position = True

square = gcode_interpreter.simulate(filename=r"first_square.gcode",printer=anisoprint_A4,initial_position=initial_position)
```

```python
square.plot_2d_position(filename=False,colvar_spatial_resolution=0.1,show=True)
```

```python
square.plot_vel(filename=False,show=True,show_plannerblocks=False,axis=("x","y"))
```
